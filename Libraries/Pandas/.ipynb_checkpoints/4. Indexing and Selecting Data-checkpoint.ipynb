```python
import pandas as pd
df = pd.read_csv("data/nba.csv")
df.head()
```

```python
name = df["Name"]
name.head(2)
```

```python
# Selecting Multiple Columns
df[["Name","Team"]]
```

```python
 # Indexing with .loc[ ]
# Selecting a Single Row by Label
df.loc[0]
```

```python
# Selecting Multiple Rows by Label
df.loc[1:3]
```

```python
# Selecting Specific Rows and Columns
df.loc[[1,3],["Name", "College"]]
```

```python
# Selecting All Rows and Specific Columns
df.loc[:,["Name",  "Team"]]
```

```python
# Indexing with .iloc[ ]
# Selecting a Single Row by Position
df.iloc[3]
```

```python
 #  Selecting Multiple Rows by Position
df.iloc[[3, 5, 7]]
```

```python
# Selecting Specific Rows and Columns by Position
df.iloc[[3, 4], [1, 2]]
```

```python
# Selecting All Rows and Specific Columns by Position
df.iloc[:, [1, 2]]
```

```python
# query(): Query the DataFrame using a boolean expression
df.query("Age > 25 and College == 'Duke'")
```
