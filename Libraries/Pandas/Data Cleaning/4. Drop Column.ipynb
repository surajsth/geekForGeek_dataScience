```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'FirstName': ['Vipul', 'Ashish', 'Milan'],
                            "Gender": ["", "", ""],
                            "Age": [0, 0, 0]})
df['Department'] = np.nan

print(df)
```

```python
 # Remove All Null Value Columns
df['Department'] = np.nan

display(df)

df.dropna(how='all', axis=1, inplace=True)

display(df)
```

```python
# Replace Empty Strings with Null and Drop Null Columns
df['Department'] = np.nan
display(df)

nan_value = float("NaN")
df.replace("", nan_value, inplace=True)

df.dropna(how='all', axis=1, inplace=True)

display(df)
```

```python
# Replace Zeros with Null and Drop Null Columns
df['Department'] = np.nan
display(df)

nan_value = float("NaN")
df.replace(0, nan_value, inplace=True)

df.dropna(how='all', axis=1, inplace=True)

display(df)
```

```python
# Replace Both Zeros and Empty Strings with Null and Drop Null Columns
df['Department'] = np.nan
display(df)

nan_value = float("NaN")
df.replace(0, nan_value, inplace=True)
df.replace("", nan_value, inplace=True)

df.dropna(how='all', axis=1, inplace=True)

display(df)
```
