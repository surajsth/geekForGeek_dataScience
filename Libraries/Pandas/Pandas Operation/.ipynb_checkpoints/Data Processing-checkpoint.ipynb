```python
# Merge Data Frames
df1 = pd.DataFrame({
        'Name':['Jeevan', 'Raavan', 'Geeta', 'Bheem'], 
        'Age':[25, 24, 52, 40], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']})
df1
```

```python
# By defining a function beforehand
# Apply function
def fun(value):
    if value > 70:
        return "Yes"
    else:
        return "No"

df['Customer Satisfaction'] = df['Spending Score (1-100)'].apply(fun)
df.head(10)
```

```python
# By using the lambda operator
const = df['Age'].max()
df['Age'] = df['Age'].apply(lambda x: x/const)
df.head()
```

```python
# Visualizing DataFrame
# scatter plot
# Visualization
df.plot(x ='CustomerID', y='Spending Score (1-100)',kind = 'scatter')
```
